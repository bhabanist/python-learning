#problem-1
==========
  The equation for the height of a throw ball is \
         $ y = v_0t -\frac{1}{2}gt^2$
         where\
         t = time\
         y = height of the ball\
         $v_0$ = initial upward speed of the ball.\
         g = gravity\
    ram throws his ball at 10 m/sec and shyam at 15 m/sec upward. Plot the height of each of their ball as a function of time.     

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,2,100)     #we defining time for which we want to observe
g = 9.81  # here we assigned the value of acceleration due to gravity
v0_r = 10
v0_s = 15
y_r = v0_r*t -0.5*g*t**2
y_s = v0_s*t -0.5*g*t**2
plt.plot(t,y_r)
plt.plot(t,y_s)
plt.xlabel("Time", fontsize = 15)
plt.ylabel("Height" ,fontsize = 15)
```

# Problem-3

Molecular dynamics

```python
# input all required ibraries
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python



# initializing all the variables
class Dynamics:
    def __init__(self):
        self.dim = 2
        self.num_particles = 50
        self.mass = 1
        self.temp = 300
        self.k = 1
        self.box_len = 10
        self.position = np.zeros((self.num_particles,self.dim))
        self.vel = np.zeros((self.num_particles,self.dim))
    def init_positions(self):       #random.sample will generate only o to 1 thats why multiply with box length
        self.positions = np.random.sample((self.num_particles,self.dim))*self.box_len
    def init_velocities(self):  
        self.std_dev = (self.mass/(self.k*self.temp))**0.5                          #random.normal will pickup vales from a normal distribution
        self.velocities = np.random.normal(loc = 0,scale = self.std_dev, size = (self.num_particles,self.dim))
    def draw_particles(self):
        plt.figure(figsize =(5,5))
        axis = plt.gca()
        axis.set_xlim(-1,self.box_len+1)
        axis.set_ylim(-1,self.box_len+1)
        
        
        for i in range(self.num_particles):
            axis.add_patch(plt.Circle(self.positions[i],radius = 0.5,linewidth = 2,edgecolor = 'black'))
            
        
```

```python
dyn = Dynamics()
dyn.init_positions()
dyn.init_velocities()

```

```python
dyn.draw_particles()
```
