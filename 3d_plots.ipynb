We will try to plot various contours. Lets try with the simplest ones.

1.$$f(x,y) = \exp(-\frac{x^2+y^2}{2\sigma^2})$$

$$-3 \leq x,y \leq 3$$

```python
import numpy as np
import matplotlib.pyplot as plt
sig= 4
x_axis = np.linspace(-3,3,100)
y_axis = np.linspace(-3,3,100)

[X,Y]=np.meshgrid(x_axis,y_axis)

f = np.exp(-(X**2+ Y**2)/(2*sig**2))


plt.contourf(X, Y, f, cmap='viridis')

# Add colorbar
plt.colorbar()


# Set plot title and labels
plt.title('Contours of Gaussian Function')
plt.xlabel('x')
plt.ylabel('y')

# Show plot
plt.show()
```

2.Next we have Rosenbrock function

$$f(x,y)=(1-x^2)+100(y-x^2)^2$$

$$-2 \leq x \leq 2 $$ and $$-1 \leq y \leq 3$$

```python
import numpy as np
import matplotlib.pyplot as plt

x_ax = np.linspace(-1.5,1.5,100)
y_ax = np.linspace(-1.5,1.5,100)

[X,Y]= np.meshgrid(x_ax,y_ax)

Z = (1-X**2) + 100*(Y-X**2)**2

#levels = np.linspace(0,1,20)
plt.contourf(X,Y,Z,cmap = "viridis")
plt.colorbar()
plt.title("Rosenbrock function")
plt.show()
```

Now we will check Booth function

$$f(x,y) = (x+2y-7)^2 + (2x+y-5)^2$$

$$0 \leq x \leq 2$$ and $$2 \leq y \leq 4 $$

```python
import numpy as np
import matplotlib.pyplot as plt

x_ax = np.linspace(-10,10,100)
y_ax = np.linspace(-10,10,100)

[X,Y]= np.meshgrid(x_ax,y_ax)

Z = (X+2*Y-7)**2 + (2*X+Y-5)**2

#levels = np.linspace(0,1,20)
plt.contourf(X,Y,Z,cmap = "viridis")
plt.colorbar()
plt.title("Booth function")
plt.show()
```

Next we have Himmelblaus funcion

$$f(x,y) = (x^2 + y - 11)^2 + (x+ y^2 - 7)^2$$

$$-5 \leq x,y \leq 5$$

```python
import numpy as np
import matplotlib.pyplot as plt

x_ax = np.linspace(-5,5,100)
y_ax = np.linspace(-5,5,100)

[X,Y]= np.meshgrid(x_ax,y_ax)

Z = (X**2+Y-11)**2 + (X+Y**2-7)**2


plt.contourf(X,Y,Z,cmap = "viridis")
plt.colorbar()
plt.title("Himmelblaus function")
plt.show()
```

Finally we will check Ackley function

$$ f(x,y) = -20 \exp(-0.2\sqrt(0.5(x^2 + y^2))) - exp(\frac{\cos(2 \pi x)+\cos(2 \pi y)}{2}) + 20 + e$$

$$-5 \leq x,y \leq 5$$

```python
import numpy as np
import matplotlib.pyplot as plt

e = 2.71
x_ax = np.linspace(-5,5,100)
y_ax = np.linspace(-5,5,100)

[X,Y]= np.meshgrid(x_ax,y_ax)

Z = -20*np.exp(-0.2*np.sqrt(0.5*(X**2 + Y**2))) - np.exp((np.cos(2*np.pi*X)+np.cos(2*np.pi*Y))/2) + 20 + e


plt.contourf(X,Y,Z,cmap = "viridis")
plt.colorbar()
plt.title("Ackley function")
plt.show()
```

```python
##2d Gaussian
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

sig= 4
x_axis = np.linspace(-10,10,100)
y_axis = np.linspace(-10,10,100)

[X,Y]=np.meshgrid(x_axis,y_axis)

f = np.exp(-(X**2+ Y**2)/(2*sig**2))

fig = plt.figure(figsize=(10,6))
ax =plt.axes(projection="3d")               
ax.contour3D(X, Y, f,50, cmap='viridis')

# Set plot title and labels
ax.set_title('Contours of Gaussian Function')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.ion()
plt.show()
```

 Same can be plotted with different styles 
 1.plot surface
 2. triangle surface
 3. wireframe

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define parameters
sig = 4
x_axis = np.linspace(-10, 10, 100)
y_axis = np.linspace(-10, 10, 100)

# Create grid of points
X, Y = np.meshgrid(x_axis, y_axis)

# Compute Gaussian function values
f = np.exp(-(X**2 + Y**2) / (2 * sig**2))

fig=plt.figure(figsize=(18,10))
# Create a new figure with subplots
#fig.add_subplot(nrows, ncols, index)

ax = fig.add_subplot(1, 3, 1, projection='3d')

# Plot wireframe
ax.plot_wireframe(X, Y, f, rstride=5, cstride=5, color='black')
ax.set_title('Wireframe')

ax = fig.add_subplot(1, 3, 2, projection='3d')
# Plot surface plot
ax.plot_surface(X, Y, f, rstride=5, cstride=5, cmap='viridis')
ax.set_title('Surface Plot')

ax = fig.add_subplot(1, 3, 3, projection='3d')
# Plot triangular surface plot

ax.plot_surface(X, Y, f, rstride=5, cstride=5,cmap='viridis', edgecolor='none')
ax.set_title('Triangular Surface Plot')



# Adjust layout and display the figure

plt.show()

```
