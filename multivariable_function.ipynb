Now lets generate functions $f(x,y) = e^{-(x^2 + y^2 )}$ for $-2 \leq x \leq 2$ and $-1 \leq x \leq 1$

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-1,1, 100)
y = np.linspace(-2,2, 100)
# now we will define meshgrid
xv,yv = np.meshgrid(x,y)
f = np.exp(-xv**2 -yv**2)
plt.pcolormesh(xv,yv,f)
plt.plot()

# Note: pcolormesh is typially preferable function for 2d plotting.
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-1,1, 100)
y = np.linspace(-2,2, 100)
# now we will define meshgrid
xv,yv = np.meshgrid(x,y)
f = np.exp(-xv**2 -yv**2)
plt.pcolormesh(xv,yv,f)
plt.colorbar()  # this will give us some sort of color with defined value
plt.plot()

#Basically its a gaussian plot so we indeed wanted this sort of figure
```

creating a reactangular mask so here $f(x,y) = 1$ for $x^2 + y^2 < 1$ and 0 for $x^2 + y^2 > 1$
for $-5 \leq x \leq 5 $ and $-5 \leq x \leq 5$

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5,5,300)
y = np.linspace(-5,5,300)
xv,yv = np.meshgrid(x,y)
f = (xv**2 + yv**2).astype(float)    
plt.pcolormesh(xv,yv,f)
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5,5,300)
y = np.linspace(-5,5,300)
xv,yv = np.meshgrid(x,y)
f = (xv**2 + yv**2 < 1).astype(float)    # here we are assigning those values for which xv^2 + yv^2 <1
plt.pcolormesh(xv,yv,f)
plt.colorbar()
plt.show()
```
